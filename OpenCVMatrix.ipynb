Transpose 90

```python
from PIL import Image
import matplotlib.pyplot as plt
import requests
from io import BytesIO

# Load the image
url = 'https://preview.redd.it/i-made-steamhappy-vector-image-v0-jmmqmwzwk14c1.png?width=800&format=png&auto=webp&s=7cc8498450fbd323b22899722ac24cbd23a91a83'
response = requests.get(url)
img = Image.open(BytesIO(response.content))

# Create a figure with two subplots
plt.figure(figsize=(10, 5))

# Display the original image
plt.subplot(121)
plt.imshow(img)
plt.title('Original Image')
plt.axis('off')

# Rotate the image 90 degrees
rotated_img = img.transpose(Image.Transpose.ROTATE_180)

# Display the rotated image
plt.subplot(122)
plt.imshow(rotated_img)
plt.title('Rotated Image (180 degrees)')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
from PIL import Image
import matplotlib.pyplot as plt
import requests
from io import BytesIO
from rembg import remove

# Load the image
url = 'https://vignette.wikia.nocookie.net/gameofthrones/images/2/25/707_Daenerys_Profil.jpg/revision/latest?cb=20170829215522&path-prefix=de'
response = requests.get(url)
img = Image.open(BytesIO(response.content))

# Create a figure with two subplots
plt.figure(figsize=(15, 7))

# Display original image
plt.subplot(121)
plt.imshow(img)
plt.title('Original Image')
plt.axis('off')

# Remove background
output = remove(img)

# Display processed image
plt.subplot(122)
plt.imshow(output)
plt.title('Background Removed')
plt.axis('off')

plt.tight_layout()
plt.show()

# Save the output if needed
output.save('daenerys_no_background.png')
```

Find 2 creative common images online. Combine these 2 images side by side vertically. Export it as Jpg

```python
from PIL import Image
import requests
from io import BytesIO

# URLs for two Creative Commons images
url1 = 'https://upload.wikimedia.org/wikipedia/commons/thumb/4/47/PNG_transparency_demonstration_1.png/240px-PNG_transparency_demonstration_1.png'
url2 = 'https://upload.wikimedia.org/wikipedia/commons/thumb/b/b2/Hausziege_04.jpg/220px-Hausziege_04.jpg'

# Download and open the images
response1 = requests.get(url1)
response2 = requests.get(url2)
img1 = Image.open(BytesIO(response1.content))
img2 = Image.open(BytesIO(response2.content))

# Convert images to RGB mode (in case they're in different modes)
img1 = img1.convert('RGB')
img2 = img2.convert('RGB')

# Calculate dimensions for the combined image
# Width will be the max of both images
total_width = max(img1.width, img2.width)
# Height will be the sum of both heights
total_height = img1.height + img2.height

# Create a new blank image with the calculated dimensions
combined_img = Image.new('RGB', (total_width, total_height), 'white')

# Paste the first image at the top
combined_img.paste(img1, (0, 0))

# Paste the second image below the first one
combined_img.paste(img2, (0, img1.height))

# Display the result
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 15))
plt.imshow(combined_img)
plt.axis('off')
plt.show()

# Save the combined image
combined_img.save('combined_images.jpg', 'JPEG', quality=95)
```

```python
from PIL import Image
import matplotlib.pyplot as plt
import requests
from io import BytesIO

# Using a Wikimedia Commons emoji URL instead
url = 'https://preview.redd.it/i-made-steamhappy-vector-image-v0-jmmqmwzwk14c1.png?width=800&format=png&auto=webp&s=7cc8498450fbd323b22899722ac24cbd23a91a83'

try:
    # Download and open the image
    response = requests.get(url)
    img = Image.open(BytesIO(response.content))

    # Create figure for displaying both images
    plt.figure(figsize=(10, 5))

    # Display original image
    plt.subplot(121)
    plt.imshow(img)
    plt.title('Original')
    plt.axis('off')

    # Convert to grayscale
    grayscale_img = img.convert('L')

    # Display grayscale image
    plt.subplot(122)
    plt.imshow(grayscale_img, cmap='gray')
    plt.title('Grayscale')
    plt.axis('off')

    plt.tight_layout()
    plt.show()

    # Save the grayscale image
    grayscale_img.save('happy_emoji_gray.png')

except Exception as e:
    print(f"An error occurred: {e}")
```
