**Q1.** randomly create 2 set of number, find the common number in 2 sets

```python
import random

# Function to generate a random set of numbers
def generate_random_set(size, min_val, max_val):
    return set(random.randint(min_val, max_val) for _ in range(size))

# Generate two random sets
set1 = generate_random_set(10, 1, 20)
set2 = generate_random_set(10, 1, 20)

# Find the common numbers (intersection)
common_numbers = set1.intersection(set2)

# Print the results
print("Set 1:", set1)
print("Set 2:", set2)
print("Common numbers:", common_numbers)
```

**Q2.** Shuffle the character in my name. for example: Tan Qi An. After shuffle nA iQ nTa

```python
import random

def shuffle_word(word):
    chars = list(word)
    random.shuffle(chars)
    return ''.join(chars)

def shuffle_name(name):
    words = name.split()
    shuffled_words = [shuffle_word(word) for word in words]
    return ' '.join(shuffled_words)

# Example usage
original_name = "Tan Qi An"
shuffled_name = shuffle_name(original_name)

print(f"Original name: {original_name}")
print(f"Shuffled name: {shuffled_name}")
```

**Truth Table**

```python
# Import pandas for better display of the table
import pandas as pd

# Define the truth values for two propositions P and Q
P_values = [True, True, False, False]
Q_values = [True, False, True, False]

# Calculate logical operations
and_values = [P and Q for P, Q in zip(P_values, Q_values)]
or_values = [P or Q for P, Q in zip(P_values, Q_values)]
not_P_values = [not P for P in P_values]

# Create a dictionary for the truth table
truth_table = {
    'P': P_values,
    'Q': Q_values,
    'P AND Q': and_values,
    'P OR Q': or_values,
    'NOT P': not_P_values
}

# Convert the dictionary into a pandas DataFrame for better visualization
df = pd.DataFrame(truth_table)

# Display the truth table
df
```

Question: take user input of proposition statement, and the output will be the truth table in dataframe format.

```python
  import pandas as pd
import itertools

# Step 1: User input for the logical expression
user_input = input("Enter a propositional logic statement (e.g., (P and Q) or not R): ")

# Step 2: Define the variables used in the expression
# For simplicity, we assume a maximum of three variables: P, Q, and R.
variables = ['P', 'Q', 'R']

# Step 3: Generate all combinations of True and False for each variable
# This gives us a list of all possible truth assignments for P, Q, and R
truth_values = list(itertools.product([True, False], repeat=len(variables)))

# Step 4: Prepare the truth table
# We will store the truth assignments and the result of the logical expression in a dictionary
truth_table = {var: [] for var in variables}  # Initialize empty lists for each variable
truth_table['Result'] = []  # This will hold the result of the logical expression

# Step 5: Evaluate the user input for each combination of truth values
for values in truth_values:
    # Create a dictionary that maps each variable to its truth value
    assignment = dict(zip(variables, values))

    # Evaluate the user input in the context of the current assignment of truth values
    result = eval(user_input, {}, assignment)

    # Add the truth values of each variable and the result to the truth table
    for var, val in assignment.items():
        truth_table[var].append(val)
    truth_table['Result'].append(result)

# Step 6: Convert the truth table into a pandas DataFrame and display it
df = pd.DataFrame(truth_table)
print("Truth Table for the expression:", user_input)
df
```
